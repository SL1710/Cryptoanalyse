```python
import requests
import pandas as pd
```

```python
def get_coin_data(coin_id, days):
    url = f"https://api.coingecko.com/api/v3/coins/{coin_id}/market_chart"
    params = {
        "vs_currency": "usd",
        "days": days
    }
    response = requests.get(url, params=params)
    data = response.json()
    
    prices = data['prices']
    market_caps = data['market_caps']
    total_volumes = data['total_volumes']
    
    timestamps = [item[0] // 1000 for item in prices]
    
    bitcoin = pd.DataFrame({'Timestamp': timestamps, 'Price': [item[1] for item in prices], 'Market Cap': [item[1] for item in market_caps], '24h Volume': [item[1] for item in total_volumes]})
    bitcoin['Timestamp'] = pd.to_datetime(bitcoin['Timestamp'], unit='s')
    bitcoin.set_index('Timestamp', inplace=True)
    
    # Pro Tag nur einen Wert erhalten
    bitcoin = bitcoin.resample('D').last()
    
    return bitcoin

coin_id = 'bitcoin'
days = 3654  # Anzahl der Tage, für die historische Daten abgerufen werden sollen

bitcoin = get_coin_data(coin_id, days)

# Extrahieren der Werte für Market Cap und 24-Stunden-Volumen ohne die Zeitstempel
bitcoin['Market Cap'] = bitcoin['Market Cap'].apply(lambda x: x[1] if isinstance(x, list) else x)
bitcoin['24h Volume'] = bitcoin['24h Volume'].apply(lambda x: x[1] if isinstance(x, list) else x)

bitcoin['Coin'] = 'BTC'
```
